```python
import re

def clean_open_text(series: pd.Series) -> pd.Series:
    return (
        series
        .fillna("")  # handle NaN safely
        .str.replace(r"[\r\n/\\;|]+", ",", regex=True)   # replace separators with comma
        .str.replace(r"\s*,\s*", ",", regex=True)        # normalize spaces around commas
        .str.replace(r",+", ",", regex=True)             # remove duplicate commas
        .str.strip(",. ")                                # strip commas, dots, spaces
    )
```

```python
df2
```

```python
df2['comfort_food']=clean_open_text(df2['comfort_food'])
```
